#### Handle missing 'Item Weight'

```python
df["Item Weight"].fillna(df["Item Weight"].mean(),inplace=True)
```

```python
df["Item Weight"].isnull().sum()
```

```python
df["Outlet Size"].isna().sum()
```

```python
df.duplicated()
```

```python
df["Item Fat Content"].value_counts()
```

#### Standardize 'Item Fat Content' values

```python
df["Item Fat Content"] = df["Item Fat Content"].replace(["LF","low fat"], "Low Fat")

df["Item Fat Content"] = df["Item Fat Content"].replace("reg", "Regular")
```

```python
df["Item Fat Content"].value_counts()
```

```python
df["Item Fat Content"] = df["Item Fat Content"].replace("Low fat", "Low Fat")
```

```python
df["Item Fat Content"].value_counts()
```

- Verified all columns have valid data types.
- Filled missing values in 'Item Weight' with mean to maintain consistency.
- Removed any duplicate entries to ensure accurate analysis.
- Standardized category labels in 'Item Fat Content' for uniformity.

## Exploratory data analysis

### Univariate Analysis

```python
df.columns
```

```python
sns.histplot(df["Sales"],kde=True)
```

```python
# Item Type frequency
plt.figure(figsize=(12,6))
df['Item Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Item Types')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Outlet Type frequency
plt.figure(figsize=(8,5))
df['Outlet Type'].value_counts().plot(kind='bar', color='coral')
plt.title('Count of Outlet Types')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='Outlet Size', palette='pastel')
plt.title('Distribution of Outlet Size')
plt.show()
```

- Analyzed distribution of 'Sales' to identify high-revenue product ranges.
- Reviewed frequency of 'Item Type' to find which categories are most common.
- Checked 'Outlet Size' and 'Outlet Type' distribution to see how store types vary.

 ## Bivariate / Multivariate Analysis

```python
# Average Sales by Item Type
plt.figure(figsize=(12,6))
df.groupby('Item Type')['Sales'].mean().sort_values().plot(kind='bar', color='green')
plt.title('Average Sales by Item Type')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Outlet Type
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Outlet Type', y='Sales')
plt.title('Sales Distribution by Outlet Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Outlet Size
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Outlet Size', y='Sales')
plt.title('Sales by Outlet Size')
plt.show()
```

```python
# Sales by Outlet Location Type
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Outlet Location Type', y='Sales')
plt.title('Sales by Outlet Location Type')
plt.show()
```

```python
# Item Visibility vs Sales
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Item Visibility', y='Sales', hue='Item Type', alpha=0.7)
plt.title('Item Visibility vs Sales')
plt.show()
```

```python
# ---------------------------------------------------------
#  6. Correlation Heatmap
# ---------------------------------------------------------

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

- Compared average sales across different 'Outlet Types' to find top-performing formats.
- Analyzed how 'Outlet Location Type' (Tier 1, Tier 2, Tier 3) impacts sales.
- Explored correlation between 'Item Visibility' and 'Sales' to see if more visible items sell better.
- Looked at 'Item Fat Content' vs. Sales to spot trends in health-focused vs. regular products.
- Analyzed the effect of 'Outlet Size' on sales performance.

## feature engineering

```python
# Convert Outlet Year to outlet age
df['Outlet Age'] = 2025 - df['Outlet Establishment Year']

# Binning Item Visibility
df['Visibility Binned'] = pd.qcut(df['Item Visibility'], q=4, labels=["Low", "Mid-Low", "Mid-High", "High"])

# Flag: High Sales
df['High Sales Flag'] = df['Sales'] > df['Sales'].median()
```

```python
df['High Sales Flag'].value_counts()
```

```python
df['Visibility Binned'].value_counts()
```

## Insights
- Supermarket Type1 outlets have higher average sales than Type2.
- Tier 1 locations tend to generate more sales than Tier 2 or Tier 3.
- High visibility items generally have slightly higher sales, suggesting better product placement can help.
- Fruits and Vegetables, Snack Foods, and Household items are top categories for revenue.
- Medium and high-sized outlets outperform small ones on average sales.
